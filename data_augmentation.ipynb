```python
def Acc(la,pr):
    r = 0
    tot = 0
    for i in range(len(la)):
        if pr[i] == la[i]:
            r += 1
            
        tot += 1
        
    return float(r)/tot
```

```python
def Predict(train_x,train_y,test_x,test_y,model,scaled):
    if scaled == 'Y':
        input_x = preprocessing.scale(train_x)
        scaler = preprocessing.StandardScaler().fit(train_x)
        output_x = scaler.transform(test_x)
        
    else:
        input_x = train_x
        output_x = test_x
    
    input_y = train_y
    model.fit(input_x,input_y)
    output_y = model.predict(output_x)
    tot = 0

    right = 0

    for i in range(len(output_y)):
        if output_y[i] == test_y[i]:
            tot +=1
            right += 1


        else:
            tot+=1

    #print(right,tot,right/tot)
    return right/tot
    

def Compare(X,Y,a,b,c,clf,scale = 'Y'):
    D = {}
    X_train = X[:a]
    Y_train = Y[:a]
    X_test_1 = X[a:b]
    Y_test_1 = Y[a:b]
    X_test_2 = X[b:c]
    Y_test_2 = Y[b:c]
    if scale == 'N':
        clf.fit(X_train,Y_train)
        y1 = clf.predict(X_test_1)
        tot1 = 0
        right1 = 0
        ID = []
        for i in range(len(y1)):
            if y1[i] == Y_test_1[i]:
                tot1 +=1
                right1 += 1
                ID.append(i)

            else:
                tot1+=1

        #print('Train(train), Test(test1)')
        #print(right1,tot1,right1/tot1)
        a1 = right1/tot1
        

        
    else:
        X_train_scaled = preprocessing.scale(X_train)
        scaler = preprocessing.StandardScaler().fit(X_train)
        X_test_1_scaled = scaler.transform(X_test_1)
        clf.fit(X_train_scaled,Y_train)
        y1 = clf.predict(X_test_1_scaled)
        tot1 = 0
        right1 = 0
        ID = []
        for i in range(len(y1)):
            if y1[i] == Y_test_1[i]:
                tot1 +=1
                right1 += 1
                ID.append(i)

            else:
                tot1+=1

        #print('Train(train), Test(test1)')
        #print(right1,tot1,right1/tot1)
        a1 = right1/tot1
        
    #D['case 1'] = a1
        
    #print('Train(train), Test(test2)')
    a2 = Predict(X_train,Y_train,X_test_2,Y_test_2,clf,scale)
    D['case 2'] = a2
    
    #print('Train(train + good test1), Test(test2)')

    X_train_2 = X[:a]
    Y_train_2 = Y[:a]

    for d in ID:
        X_train_2.append(X_test_1[d])
        Y_train_2.append(Y_test_1[d])
        
    a3 = Predict(X_train_2,Y_train_2,X_test_2,Y_test_2,clf,scale)
    D['case 3'] = a3

    #print('Train(train + noised good test1), Test(test2)')
    X_test_copy = X[a:b]
    Y_test_copy = Y[a:b]
    for d in ID:
        new_value = []
        for n in X_test_copy[d]:
            r = random.randint(-10,10)

            n = n * (100+r)/100

            new_value.append(n)
        X_test_copy[d] = new_value

    X_train_3 = X[:a]
    Y_train_3 = Y[:a]
    for d in ID:
        X_train_3.append(X_test_copy[d])
        Y_train_3.append(Y_test_copy[d])
        
    a5 = Predict(X_train_3,Y_train_3,X_test_2,Y_test_2,clf,scale)
    D['case 5'] = a5
    #print('Train(train + test1 + noised good test1), Test(test2)')
    X_train_7 = X[:b]
    Y_train_7 = Y[:b]
    for d in ID:
        X_train_7.append(X_test_copy[d])
        Y_train_7.append(Y_test_copy[d])
        
    a4 = Predict(X_train_7,Y_train_7,X_test_2,Y_test_2,clf,scale)
    D['case 4'] = a4
    #print('Train(train + test1), Test(test2)')
                       

        
    X_train_4 = X[:b]
    Y_train_4 = Y[:b]

    a6 = Predict(X_train_4,Y_train_4,X_test_2,Y_test_2,clf,scale)
    D['case 6'] = a6
    #print('Train(train + labeled(noised good test1)), Test(test2)')
    if scale == 'N':
        clf.fit(X_train,Y_train)
        
    else:
        X_train_scaled = preprocessing.scale(X_train)
        scaler = preprocessing.StandardScaler().fit(X_train)
        X_test_1_scaled = scaler.transform(X_test_1)
        clf.fit(X_train_scaled,Y_train)
        
    noised_X = []
    for d in ID:
        noised_X.append(X_test_copy[d])
        
    noised_Y = list(clf.predict(noised_X))
    X_train_5 = X[:a]
    Y_train_5 = Y[:a]
    X_train_5+=noised_X
    Y_train_5+=noised_Y

    a7 = Predict(X_train_5,Y_train_5,X_test_2,Y_test_2,clf,scale)
    D['case 7'] = a7
    #print('Train(train + test1 + labeled(noised good test1)), Test(test2)')
    X_train_6 = X[:b]
    Y_train_6 = Y[:b]
    X_train_6+=noised_X
    Y_train_6+=noised_Y

    a8 = Predict(X_train_6,Y_train_6,X_test_2,Y_test_2,clf,scale)

    D['case 8'] = a8
    
    return D
```
